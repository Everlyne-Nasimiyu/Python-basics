Guessing game

```python
import random

def guess_the_number():
  number= random.randint(1,100)
  guess  = None

  while guess != number:
    guess = int(input("Guess the number between 1 and 100: "))

    if guess < number:
      print("Too low! Try again.")
    elif guess > number:
      print("Too high! Try again")
    else:
      print("Correct! You guessed the number.")

guess_the_number()
```
